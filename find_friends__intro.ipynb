# Aplikacja znajdź znajomych

## Specyfikacja / Specs

Użytkownik ma możliwość:
* podać informacje o sobie (przedział wiekowy, płeć itp.)
* odkryć innych użytkowników, którzy są do niego podobni
* zobaczyć wspólne zainteresowania z innymi użytkownikami

## Plan / Taski
* v1 - dodawanie danych użytkownika i wyświetlanie przykładowych danych
* v2 - wykrywanie klastra do którego należy użytkownik
* v3 - dodanie opisów klastrów

## Zależności
* `streamlit` - do budowy interfejsu użytkownika.
* `openai` i modelu `gpt-4o` - do generowania opisów klastrów. 
* `dotenv` - do przechowywania klucza openai.
* `pycaret` - do wytrenowania modelu klastrowania.

## Umiejętności

* `streamlit` - moduł 6 
* `dotenv` - moduł 6
* `openai llm` - moduły 5-6
* `pycaret` - moduł 7 

## Instalujemy wymagane biblioteki (conda wymagana)

- otwórz terminal
- uruchom `conda activate od_zera_do_ai`
- uruchom `conda config --append channels conda-forge`
- uruchom `conda install -y streamlit openai python-dotenv`
